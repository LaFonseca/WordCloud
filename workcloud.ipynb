```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

texto = "A tecnologia est mudando rapidamente o mundo em que vivemos. Com a crescente popularidade da inteligncia artificial e da automao, muitos empregos esto sendo substitudos por mquinas. No entanto, a tecnologia tambm est criando novas oportunidades de emprego e melhorando a eficincia em muitas indstrias. Alm disso, a tecnologia est tornando a vida mais fcil e conveniente para as pessoas, com dispositivos inteligentes e aplicativos que ajudam a gerenciar tarefas dirias. No entanto, tambm h preocupaes sobre a privacidade e a segurana dos dados pessoais em um mundo cada vez mais conectado.  importante encontrar um equilbrio entre os benefcios e os riscos da tecnologia."


wordcloud = WordCloud(width = 1200, height = 600, 
                background_color ='black', 
                min_font_size = 10,
                contour_width = 0, contour_color = "#410F01", max_words = 1000,
                scale = 1, collocations = False, repeat = True).generate(texto) 


plt.figure(figsize = (15, 7), facecolor = None) 
plt.imshow(wordcloud)
plt.title("Top palavras do texto") 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```

```python
import requests
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import newspaper

url = "https://www.globo.com/"

article = newspaper.Article(url)

article.download()
article.parse()

content = article.text

wordcloud = WordCloud(width = 1200, height = 600, 
                background_color ='black', 
                min_font_size = 10,
                contour_width = 0, contour_color = "#410F01", max_words = 1000,
                scale = 1, collocations = False, repeat = True).generate(content)

plt.figure(figsize=(15, 7), facecolor=None)
plt.imshow(wordcloud)
plt.title("Top palavras") 
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
