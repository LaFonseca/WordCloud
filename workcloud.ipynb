```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

texto = "A tecnologia está mudando rapidamente o mundo em que vivemos. Com a crescente popularidade da inteligência artificial e da automação, muitos empregos estão sendo substituídos por máquinas. No entanto, a tecnologia também está criando novas oportunidades de emprego e melhorando a eficiência em muitas indústrias. Além disso, a tecnologia está tornando a vida mais fácil e conveniente para as pessoas, com dispositivos inteligentes e aplicativos que ajudam a gerenciar tarefas diárias. No entanto, também há preocupações sobre a privacidade e a segurança dos dados pessoais em um mundo cada vez mais conectado. É importante encontrar um equilíbrio entre os benefícios e os riscos da tecnologia."


wordcloud = WordCloud(width = 1200, height = 600, 
                background_color ='black', 
                min_font_size = 10,
                contour_width = 0, contour_color = "#410F01", max_words = 1000,
                scale = 1, collocations = False, repeat = True).generate(texto) 


plt.figure(figsize = (15, 7), facecolor = None) 
plt.imshow(wordcloud)
plt.title("Top palavras do texto") 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```

```python
import requests
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import newspaper

url = "https://www.globo.com/"

article = newspaper.Article(url)

article.download()
article.parse()

content = article.text

wordcloud = WordCloud(width = 1200, height = 600, 
                background_color ='black', 
                min_font_size = 10,
                contour_width = 0, contour_color = "#410F01", max_words = 1000,
                scale = 1, collocations = False, repeat = True).generate(content)

plt.figure(figsize=(15, 7), facecolor=None)
plt.imshow(wordcloud)
plt.title("Top palavras") 
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
